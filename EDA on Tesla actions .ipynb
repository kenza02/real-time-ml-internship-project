```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from a CSV file
df = pd.read_csv("Actions_tsla_cleaned.csv")

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the day of the week (0 = Monday, 6 = Sunday)
df['DayOfWeek'] = df['Date'].dt.dayofweek

# Calculate the average closing price for each day of the week
average_prices_by_day = df.groupby('DayOfWeek')['Close'].mean()

# Days of the week
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Convert the index of average_prices_by_day to days_of_week
average_prices_by_day.index = days_of_week

# Bar plot to visualize average closing prices by day of the week
plt.figure(figsize=(10, 6))
sns.barplot(x=days_of_week, y=average_prices_by_day)
plt.xlabel('Day of the Week')
plt.ylabel('Average Closing Price')
plt.title('Average Closing Prices by Day of the Week')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
# Summary statistics
print(df.describe())

# Line plot of closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], marker='o')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Tesla Closing Prices Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Histogram of closing prices
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Close', bins=10, kde=True)
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Tesla Closing Prices')
plt.grid(True)
plt.show()

# Box plot to visualize outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Close')
plt.xlabel('Closing Price')
plt.title('Box Plot of Tesla Closing Prices')
plt.grid(True)
plt.show()
```
