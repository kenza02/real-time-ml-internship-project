```python
from urllib.request import urlopen, Request
import os
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
finviz_url = 'https://finviz.com/quote.ashx?t=TSLA'

news_tables = {}

url = finviz_url 

req = Request(url=url, headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'})
response = urlopen(req)

html = BeautifulSoup(response, features='html.parser')
news_table = html.find(id='news-table')



parsed_data = []

for row in news_table.findAll('tr'):

        title = row.a.text
        date_time_cell = row.find('td', align='right')  # Find the cell containing date and time
        
        if date_time_cell:
            print(date_time_cell)  # Debug: Print the extracted cell
            date_time_text = date_time_cell.get_text(strip=True)
            date_time_split = date_time_text.split(' ')
            
            if len(date_time_split) == 2:
                date = date_time_split[0]
                time = date_time_split[1]
            else:
                date = None
                time = date_time_text
        else:
            date = None
            time = None
        

        parsed_data.append([date, time, title])
df = pd.DataFrame(parsed_data, columns=['date', 'time', 'title'])
output_file = 'tesla_articles.xlsx'
df.to_excel(output_file, index=False)
print(date_data)

print(f"Data has been scraped and saved to {output_file}.")

# Open the Excel file using the default program
os.system(f'start excel {output_file}')
```
